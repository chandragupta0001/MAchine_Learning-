```python
# knn
import matplotlib.pyplot as plt
from numpy import linalg as LA
import numpy as np
import pandas
train = pandas.read_csv('C:/Users/user/Desktop/datasc/iris_3/train.csv')
test = pandas.read_csv('C:/Users/user/Desktop/datasc/iris_3/test.csv')

print(test)
def sorsec(val):
    return val[0]

arry=[]
for extra in range(1,50):
    cnt=0
    for i in range(len(test)):
        a=test.iloc[i,0]
        b=test.iloc[i,1]
        rea1=test.iloc[i,2]
        rea2=test.iloc[i,3]
        rea3=test.iloc[i,4]
        
        te=[a,b,rea1,rea2]
        mat=[]
        knn=[]
        for k in range(len(train)):
            c=train.iloc[k,0]
            d=train.iloc[k,1]
            e=train.iloc[k,2]
            f=train.iloc[k,3]
            g=train.iloc[k,4]
            tra=[c,d,e,f]
            m=list(np.array(te)-np.array(tra))
            nm=LA.norm(m)
            mat.append([nm,g])
        mat.sort(key=sorsec)

        for z in range(extra):
            knn.append(mat[z][1])
        res = max(set(knn), key = knn.count)
        if(res==rea3):
            cnt+=1
    b=len(test)
    variable=(cnt/b)*100
    print(variable)
    arry.append(variable)

arry=np.array(arry)
a=[]
for k in range(1,50):
    a.append(k)
a=np.array(a)
k=max(arry)
print("maximum Accuracy for given KNN is",k)
plt.plot(a,arry,c='red')

plt.xlabel('No of neighbours : K')
plt.ylabel('Accuacy in %')
plt.grid(True)
plt.title('KNn Accuracy for Different No of Neighbours ')
plt.show()
```
