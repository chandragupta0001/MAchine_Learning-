```python
#Logistic regressionn
import matplotlib.pyplot as plt
import numpy as np
import pandas
import math
train1 = pandas.read_csv('C:/Users/user/Desktop/datasc/train.csv')
test1 = pandas.read_csv('C:/Users/user/Desktop/datasc/test.csv')
x=train1.iloc[:,0:13]
x=(x-x.mean())/(x.std())
x=x.values
ones = np.ones((x.shape[0],1),dtype=int)
x= np.concatenate((ones,x),axis=1)
y=train1.iloc[:,13:14].values
x_test=test1.iloc[:,0:13]
x_test=(x_test-x_test.mean())/(x_test.std())
x_test=x_test.values
ones = np.ones((x_test.shape[0],1),dtype=int)
x_test= np.concatenate((ones,x_test),axis=1)
y_test=test1.iloc[:,13:14].values
theta = np.zeros([1,14])
alpha = 0.01
iters = 1000
def sig(km):
    a=1/(1+math.exp(-km))
    return a
def uppsig_2(km):
    a=math.exp(-km)
    return a
arry=[]
for itera in range(1,300,10):
    theta = np.zeros([1,14])
    for i in range(itera):    
        for j in range(len(train1)):
            if(j==0):
                k=x[j]
                km = theta @ k
    #             if i==2 : print(km)
                e_fa=km[0]
                factor=((sig(e_fa)-y[j][0])*sig(e_fa)*sig(e_fa)*uppsig_2(e_fa)) * x[j]
    #             print(factor)
                df=np.array(factor)
            else :
                k=x[j]
                km= theta @ k
    #             print(km)
                e_fa=km[0]
                factor=((sig(e_fa)-y[j][0])*sig(e_fa)*sig(e_fa)*uppsig_2(e_fa)) * x[j]
    #             print(factor)
                df=np.vstack([df,factor])
        theta = theta - ((alpha)) * (np.sum(df, axis=0))

    cnt=0
    for k in range(len(test1)):
        mat=x_test[k]
        exp_pre=theta @ mat
        val=exp_pre[0]
        pre_value=sig(val)
        if(pre_value > 0.5):
            if(y_test[k][0] == 1 ):
                cnt=cnt+1
        else:
            if(y_test[k][0] == 0):
                cnt=cnt+1        
    b=len(test1)
    variable=(cnt/b)*100
    arry.append(variable)
arry=np.array(arry)
a=[]
for k in range(1,300,10):
    a.append(k)
a=np.array(a)

plt.plot(a,arry,c='red')
print("maximum accuracy when theta stablizes ",arry[10])
plt.xlabel('No of iteration')
plt.ylabel('Accuacy in %')
plt.grid(True)
plt.title('Logistic regression curve between accuracy and no of iteration')
plt.show()
```
